```python
path_root = Path("/ds2/model_zoos/zoos_sparsified/distillation/zoos")
result_root = Path("/netscratch2/jlautz/model_robustness/src/model_robustness/data/sparsified/clean_zoos")

datasets = ["MNIST", "SVHN", "CIFAR10"]
setups = ["hyp-10-f", "hyp-10-r", "seed"]
```

```python
def get_name_for_path(dataset, setup):
    
    if ds == "CIFAR10":
        size = "large"
    else:
        size = "small"
        
    if setup == "seed":
        zoo_p = f"cnn_{size}_{dataset.lower()}_ard"
    elif setup == "hyp-10-f":
        zoo_p = f"cnn_{size}_{dataset.lower()}_fixed_ard"
    else: 
        zoo_p = f"cnn_{size}_{dataset.lower()}_rand_ard"

    return zoo_p
```

```python
for ds in datasets:
    for setup in setups:
        zoo_p = get_name_for_path(ds, setup)
        checkpoint_path = os.path.join(path_root, f"{ds}/ARD/{zoo_p}")
        
        zoo_model_list = []
        for path in tqdm(os.listdir(checkpoint_path)):
            if not os.path.isfile(os.path.join(checkpoint_path, path)):
                if os.path.exists(os.path.join(checkpoint_path, path, "checkpoint_000025", "checkpoints")):
                    zoo_model_list.append(path)
        
        aux_df = pd.DataFrame() 
        zoo_model_list.sort()
        aux_df["path"] = zoo_model_list
        print(len(aux_df))
        aux_df.to_csv(os.path.join(result_root, f"{ds}_{setup}_clean_zoo.csv"))
```
